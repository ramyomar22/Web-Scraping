```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load our clean data.
df = pd.read_csv('cleaned_amazon.csv')
```

```python
# show the first rows of data.
df.head()
```

```python
# get summary statistics for data.
df.describe()
```

```python
# display data information.
df.info()
```

```python
# Check for duplicates.
df.duplicated().sum()
```

```python
# Price per GB of storage
df['price_per_gb'] = df['price'] / df['hard']
```

```python
# Price per GB of RAM.
df['price_per_ram'] = df['price'] / df['ram']
```

```python
# Number of features
num_features = df.shape[1]

# Initialize the figure
fig, axes = plt.subplots(nrows=1, ncols=num_features, figsize=(3 * num_features, 6), constrained_layout=True)

# Create a boxplot for each feature
for i, column in enumerate(df.columns):
    sns.boxplot(y=df[column], ax=axes[i])
    axes[i].set_title(f'Boxplot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Values')

# Show the plot
plt.show()
```

```python
# Plot distribution for price.
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of price')
plt.show();
```

```python
# Plot distribution for rating.
plt.figure(figsize=(10, 5))
sns.histplot(df['rating'], kde=True)
plt.title('Distribution of rating')
plt.show();
```

```python
# Plot distribution for reviews.
plt.figure(figsize=(10, 5))
sns.histplot(df['reviews'], kde=True)
plt.title('Distribution of reviews')
plt.show();
```

```python
# Plot distribution for ram.
plt.figure(figsize=(10, 5))
sns.histplot(df['ram'], kde=True)
plt.title('Distribution ram')
plt.show();
```

```python
# Plot distribution for hard.
plt.figure(figsize=(10, 5))
sns.histplot(df['hard'], kde=True)
plt.title('Distribution of hard')
plt.show();
```

```python
# Plot distribution for price per hard.
plt.figure(figsize=(10, 5))
sns.histplot(df['price_per_gb'], kde=True)
plt.title('Distribution of price_per_gb')
plt.show();
```

```python
# Plot distribution for price per ram.
plt.figure(figsize=(10, 5))
sns.histplot(df['price_per_ram'], kde=True)
plt.title('Distribution of price_per_ram')
plt.show();
```

```python
# Pairplot to visualize relationships
sns.pairplot(df)
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# removing outliers.
Q1 = df['price_per_ram'].quantile(0.25)
Q3 = df['price_per_ram'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
filtered_df = df[(df['price_per_ram'] >= lower_bound) & (df['price_per_ram'] <= upper_bound)]
```

```python
# cheek removing outliers.
filtered_df.shape
```

```python
# Plot distribution for each numeric column in filtered data
for column in filtered_df:
    plt.figure(figsize=(10, 5))
    sns.histplot(filtered_df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Pairplot to visualize relationships in filtered data
sns.pairplot(filtered_df)
plt.show()
```

```python
# Correlation heatmap for filtered data
correlation_matrix_filtered = filtered_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_filtered, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Filtered Data)')
plt.show()
```

```python
filtered_df.to_csv('amazon_filtered.csv',index=False)
```
